```python
import pandas as pd
import plotly as pl
import matplotlib.pyplot as plt

df = pd.read_csv('DataSet1_3.csv', sep = ',')
df.info()
```

```python
data = df[df['Date'].str.startswith('2015') | df['Date'].str.startswith('2014')]
data[data['Date'].str.startswith('2014')]['Volume'].sum()

# Группируем данные по годам и суммируем значения 'Volume'
volume_by_year = df.groupby(df['Date'].str[:4])['Open'].mean()

plt.figure(figsize=(10, 6))
plt.bar(volume_by_year.index, volume_by_year.values, color='blue')

plt.title('Среднее Open по годам')
plt.xlabel('Год')
plt.ylabel('Среднее Open')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
