```python
import pandas as pd
import plotly as pl
import matplotlib.pyplot as plt

df = pd.read_csv('DataSet1_3.csv', sep = ',')
df.info()
```

```python
df.head(5)
```

```python
volume_by_year = df.groupby(df['Date'].str[:4])['Open'].mean()

plt.figure(figsize=(10, 6))
plt.bar(volume_by_year.index, volume_by_year.values, color='blue')

plt.title('Среднее Open по годам')
plt.xlabel('Год')
plt.ylabel('Среднее Open')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
volume_data = [df[df['Date'].str.startswith(str(year))]['Volume']
               for year in range(2014, 2017)]

plt.figure(figsize=(10, 6))
plt.boxplot(volume_data, labels=['2014', '2015', '2016'])
plt.grid(True, axis='y', alpha=0.7)

plt.title('Боксплот диаграмма для количества транзакций с 2014 по 2016 год')
plt.xlabel('Год')
plt.ylabel('Количество транзакций')

plt.tight_layout()
plt.show()
```

```python
min_low_by_year = df.groupby(df['Date'].str[:4])['Low'].min()
years = df['Date'].str[:4].unique()

plt.figure(figsize=(10, 8))
plt.pie(min_low_by_year, labels=years, autopct='%1.1f%%', startangle=140)
plt.title('Минимальна стоимость акций за год')

plt.show()
```
